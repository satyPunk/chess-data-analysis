```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def clean_data(df):
    # Rename columns to match expected names
    df.rename(columns={
        'created_at': 'Start Time',
        'last_move_at': 'End Time',
        'turns': 'Number of Turns',
        'victory_status': 'Game Status',
        'increment_code': 'Time Increment',
        'white_id': 'White Player ID',
        'white_rating': 'White Player Rating',
        'black_id': 'Black Player ID',
        'black_rating': 'Black Player Rating',
        'moves': 'All Moves in Standard Chess Notation',
        'opening_eco': 'Opening Eco',
        'opening_name': 'Opening Name',
        'opening_ply': 'Opening Ply'
    }, inplace=True)

    # Convert time columns to datetime
    df['Start Time'] = pd.to_datetime(df['Start Time'], errors='coerce')
    df['End Time'] = pd.to_datetime(df['End Time'], errors='coerce')

    # Convert categorical data
    df['rated'] = df['rated'].astype(bool)
    df['winner'] = df['winner'].astype(str)
    df['Opening Name'] = df['Opening Name'].astype(str)

    # Convert numerical fields
    df['Number of Turns'] = pd.to_numeric(df['Number of Turns'], errors='coerce')
    df['White Player Rating'] = pd.to_numeric(df['White Player Rating'], errors='coerce')
    df['Black Player Rating'] = pd.to_numeric(df['Black Player Rating'], errors='coerce')
    df['Opening Ply'] = pd.to_numeric(df['Opening Ply'], errors='coerce')

    # Drop rows with missing values after conversion
    df = df.dropna()
    return df

def analyze_data(df):
    print("\nBasic Statistics:")
    print(df.describe())
    
    # Unique openings
    print("\nMost common openings:")
    print(df['Opening Name'].value_counts().head(10))
    
    

def main():
    # Load dataset
    df = pd.read_csv('/kaggle/input/chess/games.csv')
    df = clean_data(df)
    analyze_data(df)

if __name__ == "__main__":
    main()
```

```python
 # Rating distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['white_rating'], bins=30, kde=True, label='White Rating', color='blue')
sns.histplot(df['black_rating'], bins=30, kde=True, label='Black Rating', color='red')
plt.legend()
plt.title('Rating Distribution')
plt.show()
    
```

```python
 # Game length vs. rating
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['turns'], y=(df['white_rating'] + df['black_rating']) / 2, alpha=1, s=20)
plt.xlabel('Number of Turns')
plt.ylabel('Average Player Rating')
plt.title('Game Length vs. Player Rating')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
  # Win distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='winner', data=df, order=df['winner'].value_counts().index)
plt.title('Winner Distribution')
plt.show()
```

```python
# Opening popularity by rating
top_openings = df['opening_name'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.boxplot(x='opening_name', y=(df['white_rating'] + df['black_rating']) / 2, data=df[df['opening_name'].isin(top_openings)])
plt.xticks(rotation=90)
plt.title('Popular Openings vs. Rating')
plt.show()
```
